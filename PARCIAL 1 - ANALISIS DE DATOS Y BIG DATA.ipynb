```python
#Librerias a utilizar:
import io
import sys
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt 
import os
import time
import random
from tqdm import tqdm
from sys import getsizeof
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#Establecer ruta de los archivos y concatenarlos:
df_1 = 'ae_com.csv'
df_2 = 'amazon_com.csv'
df_3 = 'btemptd_com.csv'
df_4 = 'calvinklein_com.csv'
df_5 = 'hankypanky_com.csv'
df_6 = 'macys_com.csv'
df_7 = 'shop_nordstrom_com.csv'
df_8 = 'us_topshop_com.csv'
df_9 = 'victoriassecret_com.csv'


print("*** Merging multiple csv files into a single pandas dataframe ***")

# merge files
df = pd.concat(map(pd.read_csv, [df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9]), ignore_index=True)
print(df)
```
